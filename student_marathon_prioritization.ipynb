```python
address = input("Please enter the address: ")
street, city, state, zip, country = address.split(',')

print(f" Street: {street}\n City: {city}\n State: {state}\n Zip Code: {zip}\n Country: {country}")

"""
Students from outside the country: priority 1
Out-of-state students: priority 2
In-state students who live 100 miles or more from the University: priority 3
In-state students who live less than 100 miles from the University: priority 4
"""

priority = 0
if country.strip() != 'USA':
  priority = 1
elif state.strip() != 'FL':
  priority = 2
else:
  distance = input("Please enter the driving distance to the University (in miles): ")
  if state.strip() == 'FL' and int(distance) >= 100:
    priority = 3
  elif state.strip() == 'FL' and int(distance) < 100:
    priority = 4
print(f" Priority: {priority}")
```
